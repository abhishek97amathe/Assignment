Question  3  Using the sales data, generate a heatmap to visualize the variation in sales across different months and days

```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(20)

data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}

df4 = pd.DataFrame(data)

heatmap_data = df4.groupby(['Month', 'Day']).sum().reset_index()


fig = px.density_heatmap(heatmap_data, x='Day', y='Month', z='Sales', color_continuous_scale='Cividis',labels={'Sales': 'Total Sales'},title='Sales Heatmap')

fig.show()
```

Question 4. Using the given x and y data, generate a 3D surface plot to visualize the function z = sin(sqrt(x^2 +y^2))

```python
import plotly.graph_objects as go

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)


fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

fig.show()
```

Question 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts', toolbar_location=None, tools="")


p.vbar(x='fruits', top='counts', width=0.5, source=source, legend_field="fruits", line_color='white', fill_color='orange')

output_notebook()
show(p)
```

Question 4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)


hist_df = pd.DataFrame({'left': edges[:-1],'right': edges[1:],'top': hist,'bottom': np.zeros(len(hist))})

source = ColumnDataSource(hist_df)

p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')

p.quad(bottom='bottom', top='top', left='left', right='right', source=source, fill_color='navy', line_color='white', alpha=0.5)

output_notebook()
show(p)
```
