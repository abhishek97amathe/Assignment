```python
date_range = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')

temperature = np.random.uniform(low=-10, high=35, size=len(date_range))
humidity = np.random.uniform(low=20, high=100, size=len(date_range))
df = pd.DataFrame({
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity
})
df
```

```python

fig, ax1 = plt.subplots(figsize=(14, 7))

ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (°C)', color='b')
ax1.plot(df['Date'], df['Temperature'], color='b', label='Temperature')
ax1.tick_params(axis='y', labelcolor='b')


ax2 = ax1.twinx()
ax2.set_ylabel('Humidity (%)', color='r')
ax2.plot(df['Date'], df['Humidity'], color='r', label='Humidity')
ax2.tick_params(axis='y', labelcolor='r')



ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.title('Temperature and Humidity Over Time')
plt.show()
```

```python

import seaborn as sns

x = np.random.randn(100)
y = np.random.randn(100)


def determine_quadrant(x, y):
    if x > 0 and y > 0:
        return 'Q1'
    elif x < 0 and y > 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    elif x > 0 and y < 0:
        return 'Q4'
    else:
        return 'Origin'

quadrants = [determine_quadrant(xi, yi) for xi, yi in zip(x, y)]


df = pd.DataFrame({'x': x, 'y': y, 'quadrant': quadrants})


plt.figure(figsize=(10, 6))
scatter_plot = sns.scatterplot(data=df, x='x', y='y', hue='quadrant', palette='deep')

plt.axhline(0, color='gray', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')

plt.title('Scatter Plot of Random Points Colored by Quadrant')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(title='Quadrant')
plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()

x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

p = figure(title="Sine Wave Function",
           x_axis_label='x',
           y_axis_label='sin(x)',
           width=800,
           height=400)


p.line(x, y, legend_label="sin(x)", line_width=2)

p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'
p.xgrid.grid_line_dash = 'dashed'
p.ygrid.grid_line_dash = 'dashed'

show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Viridis256

output_notebook()

categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 101, size=len(categories))

data = pd.DataFrame({'categories': categories, 'values': values})

p = figure(x_range=categories,
           title="Random Categorical Bar Chart",
           x_axis_label='Categories',
           y_axis_label='Values',
           width=800,
           height=400,
           toolbar_location=None,
           tools="")


source = ColumnDataSource(data)
mapper = factor_cmap(field_name='categories', palette=Viridis256, factors=categories)

p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=mapper)

hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

p.title.text = 'Random Categorical Bar Chart'
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

show(p)
```
